```python
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import random
import seaborn as sn
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9)
```

```python
data = load_digits()
```

```python
df = pd.DataFrame(data.data,columns = data.feature_names)
```

```python
df['target'] = data.target
```

```python
r1 = df[df.target==0]
r1 = df[df.target==1]
r2 = df[df.target==2]
r3 = df[df.target==3]
r4 = df[df.target==4]
r5 = df[df.target==5]
r6 = df[df.target==6]
r7 = df[df.target==7]
r8 = df[df.target==8]
```

```python
plt.xlabel('pixel_0_2')
plt.ylabel('pixel_0_1')
plt.scatter(r1['pixel_0_1'],r1['pixel_0_2'],color="green",marker='+')
plt.scatter(r2['pixel_0_1'],r2['pixel_0_2'],color="blue",marker='.')
plt.scatter(r3['pixel_0_1'],r3['pixel_0_2'],color="yellow",marker='*')
plt.scatter(r4['pixel_0_1'],r4['pixel_0_2'],color="red",marker='^')
```

```python
x = df.drop(["target"],axis = 'columns' )
y = df.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
knn.fit(x_train,y_train)
```

```python
print(knn.score(x_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
